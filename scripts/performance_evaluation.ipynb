```python
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('slf4j/slf4j-timing.txt', sep='\t', header=None)
df.columns = ['PrevRev', 'Rev', 'GumTime', 'TotalTime', 'Garbage']
#df = df.reset_index()

df = df[['GumTime', 'TotalTime']]
df = df[df['GumTime'] != 0].reset_index()
df['index'] = df.index
```

```python
tmp1 = df[['index', 'GumTime']]
tmp1['Subject'] = 'GumTree'
tmp1.columns = ['index', 'time', 'Subject']
```

```python
tmp2 = df[['index', 'TotalTime']]
tmp2['Subject'] = 'Total'
tmp2.columns = ['index', 'time', 'Subject']
```

```python
mdata = pd.concat([tmp1, tmp2], axis=0)
```

```python
a4_dims = (20, 12)
fig, ax = plt.subplots(figsize=a4_dims)

# Plot tip as a function of toal bill across days
g = sns.lmplot(x="index", y="time", hue="Subject",
               truncate=True, size=10, data=mdata)
g.set_axis_labels("Commit index", "Time")
```
